```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_jobs_list(soup_GD):
    try:
        jobs_list = soup_GD.find_all('li', class_='JobsList_jobListItem__JBBUV')
    except AttributeError:
        jobs_list = []
        
    return jobs_list
```

```python
def get_comp_name(soup_GD, job):
    try:
        company_name = job.find('div', class_='css-8wag7x').text
        company_name = ''.join(i for i in company_name if ((i.isalpha()) | (i == ' ')))
        company_name = company_name.strip()
    except AttributeError:
        company_name = ''
        
    return company_name
```

```python
def get_job_location(soup_GD, job):
    try:
        job_location = job.find('div', class_='location mt-xxsm').text
    except AttributeError:
        job_location = ''
        
    return job_location
```

```python
def get_job_salary(soup_GD, job):
    try:
        estimated_salary = job.find('div', class_='salary-estimate').text.split('\xa0')[0]
    except AttributeError:
        estimated_salary = ''
        
    return estimated_salary
```

```python
URL_GD = 'https://www.glassdoor.co.in/Job/india-data-engineer-jobs-SRCH_IL.0,5_IN115_KO6,19.htm'
Hearders = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36', 'Accept-Language': 'en-US,en;q=0.5'})

webpage = requests.get(URL_GD, headers=Hearders)
company_name = []
job_location = []
estimated_salary = []
soup_GD = BeautifulSoup(webpage.content, 'html.parser')
jobs_list = get_jobs_list(soup_GD)

for job in jobs_list:
    company_name.append(get_comp_name(soup_GD, job))
    job_location.append(get_job_location(soup_GD, job))
    estimated_salary.append(get_job_salary(soup_GD, job))
    
print(company_name)
print(len(company_name))
print(job_location)
print(estimated_salary)
info_dic = {'company_name': company_name, 'job_location': job_location, 'estimated_salary': estimated_salary}
print(info_dic)
```

```python
df = pd.DataFrame.from_dict(info_dic)
```
